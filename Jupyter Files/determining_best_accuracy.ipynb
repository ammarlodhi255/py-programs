```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split as t_split
from sklearn.neighbors import KNeighborsClassifier as kneighbor


my_ds = pd.read_csv('D:\\University Documents\\Assignments\\6th Semester\\Artificial Intelligence\\Datasets\diabetes.csv')
my_ds.head()

# my_ds.iloc[0, :].values
```

```python
my_ds = my_ds[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',	'BMI',	'DiabetesPedigreeFunction', 'Age',  'Outcome']]
X = my_ds.iloc[:, :-1].values
Y = my_ds.iloc[:, 8].values

test_data = []

for testing_size in np.arange(0.1, 0.9, 0.1):
    Xtrain, Xtest, Ytrain, Ytest = t_split(X, Y, test_size=testing_size)
    for neighbors in range(1, 10, 1):
        classifier = kneighbor(n_neighbors=neighbors)
        classifier.fit(Xtrain, Ytrain)
        y_predict = classifier.predict(Xtest)
        data = [testing_size, neighbors, accuracy_score(y_predict, Ytest)]
        test_data.append(data)

print(f"test_size\tneighbors\taccuracy")
for i in range(0, len(test_data)):
    print(test_data[i])
# print(f"Accuracy is: {accuracy_score(y_predict, Ytest)}")
```
