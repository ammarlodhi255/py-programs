```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x_values = np.array([1, 2, 3])
y_values = np.array([1, 100])

plt.subplot(2, 1, 1)
plt.plot(x_values, x_values*x_values, marker = 'D')

plt.subplot(2, 1, 2)
plt.plot(x_values, np.exp(x_values))
plt.show()
```

```python
x_values = [1, 2, 6, 8]
y_values = [3, 8, 1, 10]

plt.subplot(2, 1, 1)
plt.title('Stock vs Investment', fontdict = {'family': 'serif'}, loc = 'center')
plt.plot(x_values, y_values)
plt.grid(axis = 'both', c = 'g')

plt.subplot(2, 1, 2)
plt.title('Home vs Inflation')
plt.plot(y_values, marker = 'D', ls = '--', c = 'm', lw = '1.0')
plt.plot(np.array([3, 10, 18, 20]), marker = 'o', ls = '-.', c = 'g')
plt.xlabel('Home Price')
plt.ylabel('Inflation')

plt.suptitle('Graphs')
plt.show()
```

```python
species = (my_ds['variety'].dropna()).values
setosa_num = np.count_nonzero(species == 'Setosa')
virginica_num = np.count_nonzero(species == 'Virginica')
versicolor_num = np.count_nonzero(species == 'Versicolor')
species_nums = np.array([setosa_num, virginica_num, versicolor_num])
print(f"Setosa: {setosa_num}\nVirginica: {virginica}\nVersicolor: {versicolor_num}")
plt.pie(species_nums, labels=np.array(['Setosa', 'Virginica', 'Versicolor']), startangle=90,
explode=np.array([0.5, 0.5, 0.5]), shadow = True, colors = np.array(['Black', 'Green', 'Pink', 'Yellow']))
```
